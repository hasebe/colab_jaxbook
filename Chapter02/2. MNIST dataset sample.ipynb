# MNIST Dataset Sample

Note: This notebook is designed to run on CPU (no GPU) runtime.

**[MDS-01]**

Import modules.

```
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist

plt.rcParams.update({'font.size': 12})
```

**[MDS-02]**

Download the MNIST dataset and store into NumPy arrays.

```
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()
train_images = train_images.reshape([-1, 784]).astype('float32') / 255
test_images = test_images.reshape([-1, 784]).astype('float32') / 255
train_labels = np.eye(10)[train_labels]
test_labels = np.eye(10)[test_labels]
```

**[MDS-03]**

Show the first image from the training dataset. It's a flat list of 784 pixel values.

```
train_images[0]
```

**[MDS-04]**

Show the corresponding label. In this case, the 5th (starting from 0th) element is 1. Hence, the image is "5".

```
train_labels[0]
```

**[MDS-05]**

Show some examples as images.

```
fig = plt.figure(figsize=(8, 4))
for c, (image, label) in enumerate(zip(train_images[:10], train_labels[:10])):
    subplot = fig.add_subplot(2, 5, c+1)
    subplot.set_xticks([])
    subplot.set_yticks([])
    subplot.set_title(np.argmax(label))
    subplot.imshow(image.reshape([28, 28]),
                   vmin=0, vmax=1, cmap=plt.cm.gray_r)
```
